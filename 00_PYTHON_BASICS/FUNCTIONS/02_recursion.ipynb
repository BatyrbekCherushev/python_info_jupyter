РЕКУРСІЯ

Закони рекурсії
1. Рекурсивний алгоритм повинен мати базовий випадок.  
2. Рекурсивний алгоритм повинен змінювати свій стан і рухатися до базового випадку.  
3. Рекурсивний алгоритм повинен викликати сам себе рекурсивно.

Переваги рекурсії
1. Рекурсивні функції роблять код чистим та елегантним. 
2. Складну задачу можна розбити на простіші підзадачі за допомогою рекурсії. 
3. Генерувати послідовність простіше за допомогою рекурсії, ніж за допомогою вкладених ітерацій.

Вади рекурсії:
* може використовувати більше пам'яті я часу

* інтерпретатор Python обмежує глибину рекурсії, щоб уникнути нескінченних рекурсій, що призводять до переповнення стеку.
    * За замовчуванням максимальна глибина рекурсії дорівнює **1000** . Перевищення цієї межі призводить до виникнення помилки рекурсії RecursionError .
